```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
def remove_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    threshold = 1.5 * IQR
    outlier_mask = (column < Q1 - threshold) | (column > Q3 + threshold)
    return column[~outlier_mask]
```

```python
df = df.drop_duplicates()
```

```python
col_name = ['Category','Post Month','Post Weekday','Post Hour','Paid']
for col in col_name:
    df[col] = remove_outliers(df[col])
```

```python
df.columns = df.columns.str.strip()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Type", palette="coolwarm")
plt.title("Distribution of Post Types")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Post Month", y="Lifetime Post Total Reach", marker="o", label="Total Reach")
sns.lineplot(data=df, x="Post Month", y="Lifetime Post Total Impressions", marker="s", label="Total Impressions")
plt.title("Monthly Trend of Post Reach & Impressions")
plt.xlabel("Month")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Lifetime Post Total Reach"], bins=20, kde=True, color="blue")
plt.title("Distribution of Post Reach")
plt.xlabel("Lifetime Post Total Reach")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Lifetime Post Total Impressions", y="Lifetime Engaged Users", hue="Type", alpha=0.7)
plt.title("Engagement vs. Impressions")
plt.xlabel("Total Impressions")
plt.ylabel("Engaged Users")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Type", y="Total Interactions", palette="pastel")
plt.title("Total Interactions per Post Type")
plt.xlabel("Post Type")
plt.ylabel("Total Interactions")
plt.show()
```
